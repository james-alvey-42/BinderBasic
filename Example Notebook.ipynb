## Basic Binder Example

Binder is a hosting tool for dealing with code environments such as Python and Julia where the dependencies are vital for functioning code.

This is a very basic example with a lightweight environment containing only matplotlib and numpy as well as a style file for plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('example.mplstyle')
```

```python
np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))                 
nse2 = np.random.randn(len(t))

# Two signals with a coherent part at 10Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, figsize=(6,10))
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Signal')

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel('Coherence')

plt.show()
```
